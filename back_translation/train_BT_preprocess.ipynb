```python
import re
from tqdm import tqdm

for idx in tqdm(range(len(processed_train))):
# for idx in range(40, 41):

    temp_item = processed_train.iloc[idx]
    
    temp_sentence = temp_item['sentence']
    # temp_sentence = re.sub('&[^;]*;', '', temp_sentence)
    
    trans_dict = {
        '&#39;': "'",
        '&quot;': '"',
        '&amp;': "&",
        '&lt;': "<",
        '&gt;': ">",
    }
    
    for k in trans_dict.keys():
        temp_sentence = temp_sentence.replace(k, trans_dict[k])
    
    subject_entity = eval(temp_item['subject_entity'])
    object_entity = eval(temp_item['object_entity'])
    
    s_idx = temp_sentence.find('[SUB]')
    temp_sentence = temp_sentence.replace('[SUB]', subject_entity['word'])
    subject_entity['start_idx'] = s_idx
    subject_entity['end_idx'] = s_idx + len(subject_entity['word'])
    # print(temp_sentence)
    # print(subject_entity)
    
    s_idx = temp_sentence.find('[OBJ]')
    temp_sentence = temp_sentence.replace('[OBJ]', object_entity['word'])
    object_entity['start_idx'] = s_idx
    object_entity['end_idx'] = s_idx + len(object_entity['word'])
    # print(temp_sentence)
    # print(object_entity)
    
    processed_train.iloc[idx, 1] = temp_sentence
    processed_train.iloc[idx, 2] = str(subject_entity)
    processed_train.iloc[idx, 3] = str(object_entity)
    
    
    # print(processed_train.iloc[idx])
```
